```python
import sys
import os

# Define o caminho da pasta do seu projeto
project_path = '/content/drive/MyDrive/trabalho-collab'

# Adiciona a pasta ao sys.path se ela ainda não estiver lá
if project_path not in sys.path:
    sys.path.append(project_path)

print(f'Pasta {project_path} adicionada ao path com sucesso!')
```

```python
import nbformat
from IPython import get_ipython

def executar_notebook_do_drive(caminho_relativo):
    full_path = os.path.join(project_path, caminho_relativo)
    if os.path.exists(full_path):
        print(f'\n' + '='*50)
        print(f'EXECUTANDO: {caminho_relativo}')
        print('='*50 + '\n')
        # %run executa o notebook no namespace atual, mantendo as variáveis
        get_ipython().run_line_magic('run', full_path)
    else:
        print(f'Arquivo não encontrado: {full_path}')

# Lista de módulos na ordem de execução desejada
modulos = [
    'modulos/01_perceptron/notebook.ipynb',
    'modulos/02_adaline/notebook.ipynb',
    'modulos/03_mlp/notebook.ipynb',
    'modulos/04_avaliacao/notebook.ipynb',
    'modulos/05_hebb/notebook.ipynb',
    'modulos/06_hopfield/notebook.ipynb',
    'modulos/07_kohonen/notebook.ipynb',
    'modulos/08_sem_peso/notebook.ipynb',
    'modulos/09_art/notebook.ipynb',
    'modulos/10_bam/notebook.ipynb',
    'modulos/11_cnn/notebook.ipynb'
]

# Executa cada um em sequência
for notebook_path in modulos:
    executar_notebook_do_drive(notebook_path)
```
