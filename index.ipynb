## Configuração do ambiente (recomendado)

Antes de instalar os pacotes, crie um ambiente virtual isolado para não poluir o Python global da sua máquina:

```bash
# Criar o venv (só na primeira vez)
python -m venv .venv

# Ativar — macOS/Linux
source .venv/bin/activate

# Ativar — Windows
.venv\\Scripts\\activate

# Instalar as dependências
pip install -r requirements.txt
```
